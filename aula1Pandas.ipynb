```python
#criando o dataframe
dataframe6 = pd.read_excel("AdventureWorks.xlsx")
```

```python
#tipos de dados
dataframe6.dtypes
```

```python
#receita total (valorVenda)
dataframe6['Valor Venda'].sum()
```

```python
#Qual o custo Total?
dataframe6["custo"] = dataframe6["Custo Unitário"].mul(dataframe6["Quantidade"]) #Criando a coluna de custo
```

```python
dataframe6.head(5)
```

```python
#Qual o custo Total?
round(dataframe6["custo"].sum(), 2)
```

```python
#Agora que temos a receita e custo e o total, podemos achar o Lucro total
#Vamos criar uma coluna de Lucro que será Receita - Custo
dataframe6["lucro"]  = dataframe6["Valor Venda"] - dataframe6["custo"] 
```

```python
dataframe6.head(1)
```

```python
#Total Lucro
round(dataframe6["lucro"].sum(),2)
```

```python
#Criando uma coluna com total de dias para enviar o produto
dataframe6["Tempo_envio"] = dataframe6["Data Envio"] - dataframe6["Data Venda"]
```

```python
dataframe6.head(1)
```

**Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em númerica**

```python
#Extraindo apenas os dias
dataframe6["Tempo_envio"] = (dataframe6["Data Envio"] - dataframe6["Data Venda"]).dt.days
```

```python
dataframe6.head(1)
```

```python
#Verificando o tipo da coluna Tempo_envio
dataframe6["Tempo_envio"].dtype
```

```python
#Média do tempo de envio por Marca
dataframe6.groupby("Marca")["Tempo_envio"].mean()
```
